```python
"""
BAYESIAN SPECTROSCOPIC MASS ESTIMATOR (MADE) USING A BAYESIAN NEURAL NETWORK 
WITH VRARD-APOKASC-TGAS DR14 TRAINING DATA
"""
%matplotlib inline 
from scipy.stats import kde
import scipy
import CalcBNN3Layer as cbnn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Read relevant data

```python
# Read Vrard-APOKASC-TGAS_DR14 file
#("location_id","ra","dec","vlos","evlos","j","ej","h","eh","k","ek","teff","teff_err","logg",
# "logg_err","m_h","m_h_err","alpha_m","alpha_m_err"],elarrdef,elarrdef_err,["ra_gaia",
# "dec_gaia","ra_error","dec_error","parallax","parallax_error","pmra","pmra_error","pmdec",
# "pmdec_error","meandist","diststd","mass","mass_error"])
datafile = "../data/Vrard_APOKASC_TGAS_DR14_supp_seismo_masses.csv"
star     = pd.read_csv(datafile,header=0,index_col=0) 
nstars   = len(star)      
print("The number of stars in the Vrard-APOKASC-TGAS DR14 sample is:")
print(nstars) 
print(" ")

# Read moments calculated with mass information
fitdata = ["_appmag","_col","_logg","_teff","feh","mass","parallax"]
usedata = [True,True,True,True,True,True,True]
usegalprior = True
momentsfile  = "../results/iso/moments_vrardapokasctgas"
if (usegalprior==True):
    fileinfo = "_usegalprior"
else:
    fileinfo = "_nogalprior"
if (usedata[0]==True):
    fileinfo += "_appmag"
if (usedata[1]==True):
    fileinfo += "_col"
if (usedata[2]==True):
    fileinfo += "_logg"
if (usedata[3]==True):
    fileinfo += "_teff"
if (usedata[4]==True):
    fileinfo += "_feh"
if (usedata[5]==True):
    fileinfo += "_mass"
if (usedata[6]==True):
    fileinfo += "_parallax"
momentsfile  += fileinfo+".csv"
moments = pd.read_csv(momentsfile)

# Read moments calculated without mass information
usedata = [True,True,True,True,True,False,True]
momentsfile  = "../results/iso/moments_vrardapokasctgas"
if (usegalprior==True):
    fileinfo = "_usegalprior"
else:
    fileinfo = "_nogalprior"
if (usedata[0]==True):
    fileinfo += "_appmag"
if (usedata[1]==True):
    fileinfo += "_col"
if (usedata[2]==True):
    fileinfo += "_logg"
if (usedata[3]==True):
    fileinfo += "_teff"
if (usedata[4]==True):
    fileinfo += "_feh"
if (usedata[5]==True):
    fileinfo += "_mass"
if (usedata[6]==True):
    fileinfo += "_parallax"
momentsfile  += fileinfo+".csv"
print(momentsfile)
momentsNomass = pd.read_csv(momentsfile)
                        
# Define stars without nan age/DM and those without -9999 for photospectral parameters of interest
igoodstar = ~np.isnan(moments["mu_log10age"]) & ~np.isnan(moments["sig_log10age"]) &\
            ~np.isnan(moments["mu_mh"]) & ~np.isnan(moments["sig_mh"]) &\
            ~np.isnan(moments["mu_log10mass"]) & ~np.isnan(moments["sig_log10mass"]) &\
            ~np.isnan(moments["mu_dm"]) & ~np.isnan(moments["sig_dm"]) &\
            ~np.isinf(moments["mu_log10age"]) & ~np.isinf(moments["sig_log10age"]) &\
            ~np.isinf(moments["mu_mh"]) & ~np.isinf(moments["sig_mh"]) &\
            ~np.isinf(moments["mu_log10mass"]) & ~np.isinf(moments["sig_log10mass"]) &\
            ~np.isinf(moments["mu_dm"]) & ~np.isinf(moments["sig_dm"]) &\
            ~(star["j"]==-9999) & ~(star["h"]==-9999) &\
            ~(star["k"]==-9999) & ~(star["parallax"]==-9999) &\
            ~(star["logg"]==-9999) & ~(star["teff"]==-9999) &\
            ~(star["m_h"]==-9999) & ~(star["[C/M]"]==-9999) &\
            ~(star["[N/M]"]==-9999) & ~(star["alpha_m"]==-9999) &\
            ~(star["ej"]==-9999) & ~(star["eh"]==-9999) &\
            ~(star["ek"]==-9999) & ~(star["parallax_error"]==-9999) &\
            ~(star["logg_err"]==-9999) & ~(star["teff_err"]==-9999) &\
            ~(star["m_h_err"]==-9999) & ~(star["err_[C/M]"]==-9999) &\
            ~(star["err_[N/M]"]==-9999) & ~(star["alpha_m_err"]==-9999)
print("The number of stars in the Vrard-APOKASC-TGAS DR14 sample with good masses, ages, and distances is:")
print(np.sum(igoodstar)) 
```

## Spearman correlation coefficient of spectral parameters versus mass

```python
# Select neural network inputs and errors
inputs = np.column_stack((star["h"].values,
                          star["j"].values-star["k"].values,
                          star["parallax"].values,
                          star["logg"].values,
                          star["teff"].values,
                          star["m_h"].values,
                          star["alpha_m"].values,
                          star["[C/M]"].values,
                          star["[N/M]"].values))

ninputs = len(inputs[0,:])
print("Number of inputs: ", ninputs)

errJminK = np.sqrt(star["ej"].values**2. + star["ek"].values**2.)
errInputs = np.column_stack((star["eh"].values,
                             errJminK,
                             star["parallax_error"].values,
                             star["logg_err"].values,
                             star["teff_err"].values,
                             star["m_h_err"].values,
                             star["alpha_m_err"].values,
                             star["err_[C/M]"].values,
                             star["err_[N/M]"].values))

inputs    = inputs[igoodstar,:]
errInputs = errInputs[igoodstar,:]

# Select neural network target and errors
targets    = np.column_stack((moments["mu_log10age"][igoodstar],
                              moments["mu_mh"][igoodstar],
                              moments["mu_log10mass"][igoodstar],
                              moments["mu_dm"][igoodstar]))
errTargets = np.column_stack((moments["sig_log10age"][igoodstar],
                              moments["sig_mh"][igoodstar],
                              moments["sig_log10mass"][igoodstar],
                              moments["sig_dm"][igoodstar]))
ntargets   = len(targets[0,:])
print("Number of targets: ",ntargets)
print(" ")

labels = ["H","J-K","parallax","log g",r"$T_{\mathrm{eff}}$","[M/H]","[a/M]","[C/M]","[N/M]"]

# Calculate Spearman correlation coefficients for log10age
spearman = np.zeros([ninputs,2])
for jvar in range(ninputs):
    x = inputs[:,jvar]
    y = moments["mu_log10age"][igoodstar]
    spearman[jvar,:] = scipy.stats.spearmanr(x,y)  
print("Spearman rho and p-values for log10age:")
print(spearman)
print(" ")

# Calculate Spearman correlation coefficients for [M/H]
spearman = np.zeros([ninputs,2])
for jvar in range(ninputs):
    x = inputs[:,jvar]
    y = moments["mu_mh"][igoodstar]
    spearman[jvar,:] = scipy.stats.spearmanr(x,y)  
print("Spearman rho and p-values for [M/H]:")
print(spearman)
print(" ")

# Calculate Spearman correlation coefficients for log10mass
spearman = np.zeros([ninputs,2])
for jvar in range(ninputs):
    x = inputs[:,jvar]
    y = moments["mu_log10mass"][igoodstar]
    spearman[jvar,:] = scipy.stats.spearmanr(x,y)  
print("Spearman rho and p-values for log10mass:")
print(spearman)
print(" ")

# Calculate Spearman correlation coefficients for distance modulus
spearman = np.zeros([ninputs,2])
for jvar in range(ninputs):
    x = inputs[:,jvar]
    y = moments["mu_dm"][igoodstar]
    spearman[jvar,:] = scipy.stats.spearmanr(x,y)  
print("Spearman rho and p-values for dm:")
print(spearman)
```

# Build neural network

## Create training set and independent testing set

```python
# Split intro testing and training samples with set random seed
np.random.seed(10)
ftest           = 0.2
nobs            = inputs.shape[0]
itest           = np.random.randint(0,nobs-1,np.int(nobs*ftest))
booltest        = np.zeros([nobs],dtype=bool)
booltest[itest] = True
inputsTest      = inputs[booltest,:]
errInputsTest   = errInputs[booltest,:]
targetsTest     = targets[booltest,:]
errTargetsTest  = errTargets[booltest,:]
inputsTrain     = inputs[~booltest,:]
errInputsTrain  = errInputs[~booltest,:]
targetsTrain    = targets[~booltest,:]
errTargetsTrain = errTargets[~booltest,:]

ntrain = len(inputsTrain)
ntest  = len(inputsTest)
print("Number of training samples: ", ntrain)
print("Number of testing samples: ", ntest)
```

## Train neural network

```python
# Instantiate class
bnn = cbnn.CalcBNN3Layer()

# Number of neurons per hidden layer
neuronsPerHiddenlayer = 10

# Whether to use NUTS or ADVI sampler
sampler               = "nuts"

# Number of samples tro draw from posterior
nsamp                 = 10000

# Pickle file for storing model
tracedir = "../results/bnn/"
bnnmodelpkl = tracedir+"bnn_"+sampler+"_ninputs"+str(ninputs)+"_nhid"+str(neuronsPerHiddenlayer)+"_nsamp"+str(nsamp)+".pkl"
print(bnnmodelpkl)

# Directory for storing plots
plotdir     = "../plots/bnn/"
ncores      = 3

# Train BNN
bnn.trainBNN(inputsTrain,errInputsTrain,targetsTrain,errTargetsTrain,
             neuronsPerHiddenlayer,sampler,nsamp,bnnmodelpkl,plotdir,
             ncores,viewBNN=False)
```

## Calculate posterior predictive distributions for training set

```python
# Number of posterior predictive samples
nppc = 1000  

# Apply model
makePlots  = True
plotdir     = "plots/bnn/train/"
plotlabels = [r"$\log_{10}\tau$", "[M/H]", r"$\log_{10}m$",r"$\mu$"]

out = bnn.calcPostPredSamplesTest(bnnmodelpkl,inputsTrain,errInputsTrain,
                                  targetsTrain,errTargetsTrain,
                                  nppc,makePlots,plotlabels,plotdir)

inputsTrainTrueMu   = out[0]["InputsTrueMu"]
inputsTrainTrueSig  = out[0]["InputsTrueSig"]
targetsTrainTrueMu  = out[0]["TargetsTrueMu"]
targetsTrainTrueSig = out[0]["TargetsTrueSig"]
```

## Calculate posterior predictive distributions for testing set

```python
plotdir     = "plots/bnn/test/"
outTest = bnn.calcPostPredSamplesTest(bnnmodelpkl,inputsTest,errInputsTest,
                                      targetsTest,errTargetsTest,
                                      nppc,makePlots,plotlabels,plotdir)

inputsTestTrueMu   = outTest[0]["InputsTrueMu"]
inputsTestTrueSig  = outTest[0]["InputsTrueSig"]
targetsTestTrueMu  = outTest[0]["TargetsTrueMu"]
targetsTestTrueSig = outTest[0]["TargetsTrueSig"]
```
